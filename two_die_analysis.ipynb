```python
ev_two_die_prob = sum([a*b for a,b in zip(x_axis, two_die_prob)])
ev_two_die_prob_with_reroll = sum([a*b for a,b in zip(x_axis, two_die_prob_with_reroll)])
print ("expected value of rolling two dies = {}".format(ev_two_die_prob))
print ("expected value of rolling two dies and then rerolling the lesser = {}".format(ev_two_die_prob_with_reroll))
```

#Conclusion
